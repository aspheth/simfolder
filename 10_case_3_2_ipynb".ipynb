```python
with open ('entries.csv','r') as ef:
  entrlst = ef.readlines()
```

```python
with open('registrations.csv','r') as rf:
  reglst = rf.readlines()
```

```python
regs = []
for line in reglst:
    if '2021-01' in line and 'user_id' not in line:  # пропускаем заголовок
        user_id, date_str = line.strip().split(';')
        regs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
```

```python
entrs = []
for line in entrlst:
  if 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        entrs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
```

```python
rolling_retention = 0
entrs = set(entrs)
for x in regs:
  for y in entrs:
    if x[0] == y[0] and (y[1] - x[1]).days >= 30:
      rolling_retention += 1
```

```python
rolling_retention = round((rolling_retention / len(regs))*100,5)
```
