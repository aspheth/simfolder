### **Посчитайте Rolling-retention 30 дня (в процентах) для пользователей из той же когорты**

Сохраните результат в переменную `rolling_retention`

**Примечание:** результат округлите до 5 знаков после запятой

```python
# Ваше решение

from datetime import datetime
with open ('entries.csv','r') as ef:
  entrlst = ef.readlines()
with open('registrations.csv','r') as rf:
  reglst = rf.readlines()
regs = []
for line in reglst:
    if '2021-01' in line and 'user_id' not in line:  # пропускаем заголовок
        user_id, date_str = line.strip().split(';')
        regs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
entrs = []
for line in entrlst:
  if 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        entrs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
rolling_retention = 0
entrs = set(entrs)
for x in regs:
  for y in entrs:
    if x[0] == y[0] and (y[1] - x[1]).days >= 30:
      rolling_retention += 1
rolling_retention = round((rolling_retention / len(regs))*100,5)
```

```python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
correct_answer = float(answers[1])

try:
    assert rolling_retention == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```

### **Посчитайте Churn rate 29 дня (в долях), посчитанный по всем пользователям**

Сохраните результат в переменную `churn_29`.  
Если будете считать CR от Retention, ведите расчет от Rolling Retention, таким образом, мы получим всех, кто не заходил в 29 день и после. Ведя расчет от обычного Retention, наоборот - получим только CR в 29 день.


```python
# Ваше решение

from datetime import datetime
with open ('entries.csv','r') as ef:
  entrlst = ef.readlines()
with open('registrations.csv','r') as rf:
  reglst = rf.readlines()
regs = []
for line in reglst:
    if 'user_id' not in line:  # пропускаем заголовок
        user_id, date_str = line.strip().split(';')
        regs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
entrs = []
for line in entrlst:
  if 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        entrs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
rolling_retention = 0
entrs = set(entrs)
for x in regs:
  for y in entrs:
    if x[0] == y[0] and (y[1] - x[1]).days == 29:
      rolling_retention += 1
      break
rolling_retention = rolling_retention / len(regs) if len(regs) > 0 else 0

# Churn на 29-й день в долях
churn_29 = 1 - rolling_retention
```

```python
from datetime import datetime

with open('entries.csv', 'r') as ef:
    entrlst = ef.readlines()
with open('registrations.csv', 'r') as rf:
    reglst = rf.readlines()

regs = []
for line in reglst:
    if '2021-01' in line and 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        regs.append((int(user_id), datetime.strptime(date_str, '%Y-%m-%d').date()))

entrs = []
for line in entrlst:
    if 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        entrs.append((int(user_id), datetime.strptime(date_str, '%Y-%m-%d').date()))

entrs = list(set(entrs))  # убираем дубликаты

dau_sum = 0  # сумма DAU за все дни

for day in range(1, 32):
    current_date = datetime(2021, 12, day).date()
    dau_day = 0

    for uid, event_date in entrs:
        if event_date == current_date:
            dau_day += 1

    dau_sum += dau_day

avg_dau = round(dau_sum / 31, 5)

```
