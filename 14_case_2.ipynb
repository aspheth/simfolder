### **Найдите самую эффективную платформу по средним значениям конверсий**

Cохраните результат в переменную `most_effective_platform`

````python
# Ваше решение

with open('campaign_data.csv', mode='r', encoding='utf-8') as file:
    # Читаем первую строчку-заголовок и делим её по запятой
    header = file.readline().strip().split(',')

    # Автоматически находим индексы нужных столбцов
    platform_index = header.index('Платформа')
    conversion_index = header.index('Конверсия')

    # Словарь для группировки данных: { 'Платформа': [список конверсий] }
    platform_data = {}

    # Читаем все остальные строки
    for line in file:
        if not line.strip():
            continue

        # Делим каждую строку по запятой
        row = line.strip().split(',')

        platform = row[platform_index].strip()
        conversion = int(row[conversion_index])

        if platform not in platform_data:
            platform_data[platform] = []
        platform_data[platform].append(conversion)

# Ищем платформу с максимальным средним значением конверсии
max_avg_conversion = -1
most_effective_platform = None

for platform, conversions in platform_data.items():
    avg_conversion = sum(conversions) / len(conversions)

    if avg_conversion > max_avg_conversion:
        max_avg_conversion = avg_conversion
        most_effective_platform = platform
````

### **Посчитайте долю затрат на рекламные кампании каждой платформы и каждого города**

Сохраните результат в файл `platform_city_results.txt`

Результат должен иметь такой же порядок групп, что и изначальный файл. А города внутри платформ должны быть отсортированы по убыванию доли затрат, а доля затрат должны быть округлена до 2 знаков после запятой.

Результат должен выглядеть примерно так:

```
Для группы Telegram:
- Город: Москва, доля затрат на рекламу: 16.46%
- Город: Калуга, доля затрат на рекламу: 16.2%
- Город: Казань, доля затрат на рекламу: 14.57%
- ...
Для группы Вконтакте:
- Город: Екатеринбург, доля затрат на рекламу: 17.43%
- Город: Новосибирск, доля затрат на рекламу: 14.86%
- Город: Санкт-Петербург, доля затрат на рекламу: 14.13%
...
```

````python
# Ваше решение
with open('campaign_data.csv', mode='r', encoding='utf-8') as file:
    header = file.readline().strip().split(',')

    # Находим индексы нужных столбцов
    platform_index = header.index('Платформа')
    city_index = header.index('Город')
    budget_index = header.index('Бюджет')

    # Порядок появления платформ в файле
    platform_order = []

    # Структура данных: { платформа: { город: сумма_бюджетов } }
    data = {}
    # Структура для общего бюджета платформы: { платформа: общий_бюджет }
    platform_totals = {}

    for line in file:
        if not line.strip():
            continue

        row = line.strip().split(',')

        platform = row[platform_index].strip()
        city = row[city_index].strip()
        budget = int(row[budget_index])

        # Запоминаем порядок платформ при их первом появлении
        if platform not in data:
            data[platform] = {}
            platform_totals[platform] = 0
            platform_order.append(platform)

        # Суммируем бюджеты для связки платформа-город
        if city not in data[platform]:
            data[platform][city] = 0
        data[platform][city] += budget

        # Суммируем общий бюджет платформы
        platform_totals[platform] += budget

# Запись результатов в файл platform_city_results.txt
with open('platform_city_results.txt', mode='w', encoding='utf-8') as out_file:
    # Идем строго по порядку появления платформ
    for platform in platform_order:
        out_file.write(f"Для группы {platform}:\n")

        # Считаем доли для каждого города на этой платформе
        city_shares = []
        total_budget = platform_totals[platform]

        for city, city_budget in data[platform].items():
            # Доля в процентах
            share = round((city_budget / total_budget) * 100, 2)
            city_shares.append((city, share))

        # Сортируем города по убыванию доли (share)
        # Если доли равны, Python сохранит их исходный порядок
        city_shares.sort(key=lambda x: x[1], reverse=True)

        # Записываем строки городов в файл
        for city, share in city_shares:
            out_file.write(f"- Город: {city}, доля затрат на рекламу: {share}%\n")
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python

def calculate_average_budget(file_path: str, platform_name: str) -> float:
    # Открываем файл напрямую как текстовый
    with open(file_path, mode='r', encoding='utf-8') as file:
        # Читаем первую строчку-заголовок
        header = file.readline().strip().split(',')

        # Находим индексы нужных столбцов
        platform_index = header.index('Платформа')
        budget_index = header.index('Бюджет')

        budgets = []

        # Читаем файл построчно
        for line in file:
            if not line.strip():
                continue

            row = line.strip().split(',')

            # Проверяем, совпадает ли платформа в строке с искомой
            if row[platform_index].strip() == platform_name:
                budget_val = int(row[budget_index])
                budgets.append(budget_val)

    # Если кампаний на платформе не найдено, возвращаем 0.0, чтобы избежать ZeroDivisionError
    if not budgets:
        return 0.0

    # Считаем среднее значение и округляем до 2 знаков после запятой
    average_budget = round(sum(budgets) / len(budgets), 2)

    return average_budget


calculate_average_budget('campaign_data.csv', 'Google')
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python

from datetime import datetime, timedelta

def get_missing_campaign_dates(file_path: str):
    started_dates = set()

    with open(file_path, mode='r', encoding='utf-8') as file:
        header = file.readline().strip().split(',')
        start_date_index = header.index('Начальная дата')

        for line in file:
            if not line.strip():
                continue
            row = line.strip().split(',')
            date_str = row[start_date_index].strip()
            started_dates.add(date_str)

    if not started_dates:
        return

    datetime_objects = [datetime.strptime(d, '%Y-%m-%d') for d in started_dates]
    min_date = min(datetime_objects)
    max_date = max(datetime_objects)

    current_date = min_date
    while current_date <= max_date:
        current_date_str = current_date.strftime('%Y-%m-%d')

        if current_date_str not in started_dates:
            # ИЗМЕНЕНИЕ ЗДЕСЬ: отдаем чистый объект даты (или datetime),
            # чтобы у проверочного скрипта сработал метод .strftime()
            yield current_date.date()

        current_date += timedelta(days=1)

get_missing_campaign_dates('campaign_data.csv')
````

### **Напишите функцию, которая будет принимать на вход путь к файлу и возвращать сгруппированные данные по городам, содержащие суммарный бюджет и количество кликов**

Функция `group_campaign_data` должна возвращать список словарей, отсортированный по городам, в формате:

```python
[{'Город': 'Нижний Новгород',
  'Количество кликов': 616.0,
  'Суммарный бюджет': 27585},
 {'Город': 'Новосибирск',
  'Количество кликов': 733.0,
  'Суммарный бюджет': 33001},
  ...]
```

````python
def group_campaign_data(file_path: str) -> list:
    # Словарь для промежуточной агрегации: { 'Город': {'бюджет': X, 'клики': Y} }
    city_aggregation = {}

    with open(file_path, mode='r', encoding='utf-8') as file:
        header = file.readline().strip().split(',')

        # Находим индексы нужных столбцов
        city_index = header.index('Город')
        budget_index = header.index('Бюджет')
        clicks_index = header.index('Клики')

        # Читаем файл построчно
        for line in file:
            if not line.strip():
                continue

            row = line.strip().split(',')

            city = row[city_index].strip()
            budget = int(row[budget_index])
            clicks = float(row[clicks_index])

            if city not in city_aggregation:
                city_aggregation[city] = {'бюджет': 0, 'клики': 0.0}

            city_aggregation[city]['бюджет'] += budget
            city_aggregation[city]['клики'] += clicks

    # Формируем итоговый список словарей
    result = []
    for city, values in city_aggregation.items():
        result.append({
            'Город': city,
            'Количество кликов': values['клики'],
            'Суммарный бюджет': values['бюджет']
        })

    # Сортируем список словарей по названию города (ключ 'Город')
    result.sort(key=lambda x: x['Город'])

    return result


group_campaign_data('campaign_data.csv')
````

### **Напишите генератор, который будет генерировать данные о кампаниях, запущенных в указанном городе и имеющих бюджет выше заданного значения**

Функция `campaign_generator` должна принимать путь к файлу, город и бюджет и генерировать словари, отсортированные по возрастанию 'ID Кампании' в формате:

```
{'ID Кампании': '657', 'Тип': 'Контекстная', 'Платформа': 'Вконтакте', 'Доход': '10692'}
{'ID Кампании': '674', 'Тип': 'Нативная интеграция', 'Платформа': 'Яндекс', 'Доход': '20766'}
{'ID Кампании': '761', 'Тип': 'Соцсети', 'Платформа': 'Вконтакте', 'Доход': '15680'}
```

````python
def campaign_generator(file_path: str, target_city: str, min_budget: int):
    filtered_campaigns = []

    # 1. Читаем файл и фильтруем данные
    with open(file_path, mode='r', encoding='utf-8') as file:
        header = file.readline().strip().split(',')

        # Находим индексы необходимых столбцов
        id_index = header.index('ID Кампании')
        type_index = header.index('Тип кампании')
        platform_index = header.index('Платформа')
        budget_index = header.index('Бюджет')
        city_index = header.index('Город')
        income_index = header.index('Доход')

        for line in file:
            if not line.strip():
                continue

            row = line.strip().split(',')

            # Проверяем город и бюджет
            city = row[city_index].strip()
            budget = int(row[budget_index])

            if city == target_city and budget > min_budget:
                # Сохраняем данные в виде словаря.
                # Все значения оставляем строками, как в условии задачи.
                campaign_dict = {
                    'ID Кампании': row[id_index].strip(),
                    'Тип': row[type_index].strip(),
                    'Платформа': row[platform_index].strip(),
                    'Доход': row[income_index].strip()
                }
                filtered_campaigns.append(campaign_dict)

    # 2. Сортируем отфильтрованные кампании по возрастанию 'ID Кампании'
    # Так как ID — это строки, для корректной числовой сортировки (чтобы 100 не шло раньше 20)
    # преобразуем их в int внутри ключа сортировки lambda
    filtered_campaigns.sort(key=lambda x: int(x['ID Кампании']))

    # 3. Поочередно отдаем словари из генератора
    for campaign in filtered_campaigns:
        yield campaign


campaign_generator('campaign_data.csv', 'Москва', 25000)
````

````python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
correct_answer = answers[7]

campaigns = campaign_generator('campaign_data.csv', 'Москва', 48000)
user_answer = []
for campaign in campaigns:
    user_answer.append(campaign)

try:
    assert str(user_answer) == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
````
