```python
with open ('campaign_data.csv', 'r') as f:
  lines = f.readlines()
```

```python
lst = []
for x in lines[1:]:

  lst.append(x.split(','))
```

```python
# лучшая платформа по средним конверсии
```

```python
lst[:2]
```

```python
comps = set()
for x in lst:
  comps.add(x[4])
```

```python
list(comps)
```

```python
sum = 0
ln = 0

for x in list(comps):
  for y in lst:
    if x in y:
      sum += int(y[9])
      ln += 1
  print(f'{x} {sum/ln}')
```

```python
# лучший город по средним конверсии
```

```python
cities = set()
for x in lst:
  cities.add(x[6])
```

```python
list(cities)
```

```python
sum = 0
ln = 0

for x in list(cities):
  for y in lst:
    if x in y:
      sum += int(y[9])
      ln += 1
  print(f'{x} {sum/ln}')
```
