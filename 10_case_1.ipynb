```python
import os
import re

def process_files(src_folder, dest_folder):
    # 1. Список файлов в src_folder
    files = os.listdir(src_folder)
    matches_files = []

    for x in files:
        if re.search(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv', x):
            matches_files.append(os.path.join(src_folder, x))  # полный путь!

    if not matches_files:
        print("Нет подходящих файлов")
        return

    # 2. Объединяем данные
    resultlist = []
    header = None

    for filename in matches_files:
        with open(filename, 'r') as f:
            lines = f.readlines()
            if header is None:
                header = lines[0].strip()  # сохраняем заголовок
            for x in lines[1:]:           # пропускаем заголовок
                x = x.strip()
                x = re.sub(r'^\d+;', '', x)  # очистка
                resultlist.append(x)

    # 3. Сортируем по дате (первые 10 символов = дата)
    resultlist.sort(key=lambda line: line[:10])

    # 4. Записываем результат
    os.makedirs(dest_folder, exist_ok=True)
    dest_file = os.path.join(dest_folder, 'combined_data.csv')

    with open(dest_file, 'w') as f:
        f.write(header + '\n')
        for line in resultlist:
            f.write(line + '\n')

    print(f"Готово! Файл: {dest_file}")
```

```python
src_folder = 'reports-main'
dest_folder = 'comb_reports'
process_files(src_folder, dest_folder)
```
