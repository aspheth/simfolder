```python
#Python. ДЗ 5: Ограничить число границами диапазона
```

```python
def lim_range(num,n_min,n_max):
    if num < n_min:
        return n_min
    elif num > n_max:
        return n_max
    else: 
        return num 
```

```python
lim_range(7.3,5,10)
```

```python
#Python ДЗ 4: Полная стоимость заказа
```

```python
items = [
        {'name': 'T-shirt', 'price': 20, 'quantity': 2},
        {'name': 'Jeans', 'price': 50, 'quantity': 1},
        {'name': 'Sneakers', 'price': 80, 'quantity': 1},
    ]
```

```python
def order_total(items):
    sum = 0
    for x in items:
        sum += (x['price'] * x['quantity'])
    return sum    
```

```python
order_total(items)
```

```python
#Python ДЗ 4: Сумма покупки после скидок
```

```python
def calculate_purchase(good_type,qty,prc):
    discount = 1
    if good_type == 'A' and qty > 500:
        discount = 0.05
    elif good_type == 'B' and qty > 1000:
        discount = 0.1
    elif good_type == 'C' and qty > 2000:  
        discount = 0.15  
    return (qty * prc) - (qty * prc)*discount
```

```python
calculate_purchase("B", 1500, 20.5)
```

```python
calculate_purchase("A", 600, 10)
```

```python
#Python ДЗ 4: Штраф библиотеки
```

```python
def calculate_fine(days):
    fine = 0
    if days <= 5:
        fine = 0.5 * days   
    elif days <= 10:
        fine = 0.5 * 5 + (days - 5) 
    else:
        fine =  0.5 * 5 + 5 + (days - 10)* 5
    return fine    

```

```python
calculate_fine(24)
```

```python
def steel_assessment(hs,cc,ts):
    if hs > 50 and cc < 0.7 and ts > 5600:
     return 10
    elif hs > 50 and cc < 0.7:
     return 9
    elif cc < 0.7 and ts > 5600:
     return 8
    elif hs > 50 and ts > 5600:
     return 7
    elif hs > 50 or cc < 0.7 or ts > 5600:
     return 6
    else:
     return 5
    
```
