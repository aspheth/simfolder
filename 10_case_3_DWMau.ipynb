```python
#Посчитайте Mau, Wau, Dau за последний месяц/неделю/день записей
#Сохраните результат в переменные dec_mau, dec_wau, dec_dau соответственно

#Примечание: последний месяц записей - декабрь. Поэтому mau рассчитываем для декабря (2021 года), для wau берем последнюю неделю - с 25 по 31 декабря, и для dau соответственно последний день - 31 декабря.
```

```python
from datetime import datetime
with open ('entries.csv','r') as ef:
  entrlst = ef.readlines()
with open('registrations.csv','r') as rf:
  reglst = rf.readlines()
regs = []
for line in reglst:
    if '2021-01' in line and 'user_id' not in line:  # пропускаем заголовок
        user_id, date_str = line.strip().split(';')
        regs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
entrs = []
for line in entrlst:
  if 'user_id' not in line:
        user_id, date_str = line.strip().split(';')
        entrs.append((int(user_id), (datetime.strptime(date_str, '%Y-%m-%d')).date()))
entrs = set(entrs)
entrs = list(entrs)
edgedate = datetime(2021, 12, 31).date()
dec_mau = []
for x in entrs:
  if datetime(2021, 12, 1).date() <= x[1] <= edgedate:
    dec_mau.append(x[0])
dec_mau = len(set(dec_mau))
```
