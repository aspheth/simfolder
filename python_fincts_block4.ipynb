```python
def count_banknotes(amount, num_100_available, num_50_available, num_10_available):
     num_100 = min(amount // 100, num_100_available)
     remaining_amount = amount - num_100 * 100
     num_50 = min(remaining_amount // 50, num_50_available)
     remaining_amount -= num_50 * 50
     num_10 = min(remaining_amount // 10, num_10_available)
     return (num_100, num_50, num_10)
```
