```python
grades = [3, 5, 3, 2, 3, 4, 5]
fives = grades.count(5)
threes = grades.count(3)

f'На этой неделе было {round(threes * 100/len(grades),2)}% троек и {round(fives * 100/len(grades),2)}% пятерок'
```

```python
#Python ДЗ 1: Обновление корзины
```

```python
cart = []
a = 'grill' 

b = ['garlic sauce', 'sweet mustard sauce'] 

c = 'meat'
```

```python
cart1
```

```python
cart2 = cart1.copy()
```

```python
#Python ДЗ 1: Посетители банка
```

```python
visitors1 = (45, 10, 39, 51, 42, 38, 49)
visitors2 = (65, 14, 33, 11, 43, 44, 12)
visitors3 = (44, 45, 22, 15, 43, 22, 78)
visitors4 = (43, 33, 11, 7, 34, 31, 49)
```

```python
f'минимальное количество посетителей: {min(visitors1 + visitors2 + visitors3 + visitors4)}'
```

```python
f'максимальное количество посетителей: {max(visitors1 + visitors2 + visitors3 + visitors4)}
```

```python
f'суммарное число посетителей: {sum(visitors1 + visitors2 + visitors3 + visitors4)}'
```

```python
res = ((visitors1 + visitors2 + visitors3 + visitors4).index(max(visitors1 + visitors2 + visitors3 + visitors4)))
```
