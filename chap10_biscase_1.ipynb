```python
matches_files = []
for x in files:
  matches = re.findall(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv',x)
  if matches:
    matches_files.append(matches)
```

```python
import re
pattern = re.compile(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv')
result = [name for name in files if pattern.match(name)]
# или если возможны пробелы в начале: pattern.search(name)
```
