```python
matches_files = []
for x in files:
  matches = re.findall(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv',x)
  if matches:
    matches_files.append(matches)
```

```python
import re
pattern = re.compile(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv')
result = [name for name in files if pattern.match(name)]
# или если возможны пробелы в начале: pattern.search(name)
```

```python
import re

resultlist = []
for filename in result:
    with open(filename, 'r') as f:
        lines = f.readlines()
        for x in lines[1:]:                     # пропускаем заголовок
            x = re.sub(r'^\d+;', '', x)         # удаляем "0;", "1;" и т.д.
            resultlist.append(x)
```

```python
sorted(resultlist)
```

```python
with open("combined_data.csv",'w') as cd:
  with open(result[0],'r') as r:
    header = r.readline()
    cd.writelines(header)
    cd.writelines(sorted(resultlist))

```
