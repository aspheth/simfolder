```python
lst = [2,3,5,6,4,56,32,4,1,6,12]
```

```python
def divisible_by_2_and_3(lst):
  sum = 0
  for x in lst:
    if x % 2 == 0 and x % 3 == 0:
      sum += x
  return sum
```

```python
divisible_by_2_and_3(lst)
```

```python
#Python ДЗ 5: Наименьший делитель
```

```python
def smallestDivisor(n):
  delit = 2
  del_lst = []
  while delit < n:
    if n % delit ==0:
      del_lst.append(delit)
    delit+=1
  if del_lst:
    return min(del_lst)
  else:
    return n
```

```python
smallestDivisor(7)
```

```python
#Python ДЗ 5: Проверка на пустые строки
```

```python
def adder(lst):
  res = []
  if lst == []:
    return 'Список пуст'
  else:
    for x in lst:
      if x == '':
        return ('Есть пустая строка', res)
      else:
        res.append(x)
  return res
```

```python
adder(lst)
```

```python
#Python ДЗ 5: Сумма натуральных чисел
```

```python
def find_n(n):
  res = 0
  count = 0
  while res < n:
    count += 1
    res += count
  return count
```

```python
find_n(4)
```

```python
#Python ДЗ 5: Поисковые запросы для рекламы
```

```python
lst = ["phone case", "laptop", "headphones"]
```

```python
def add_rank(lst,rank_start = 1):
  res = []
  for x in lst:
    res.append(f'{x}#{rank_start}')
    rank_start +=1
  return res


```

```python
add_rank(lst,3)
```

```python
new_lst = []
for x,y in enumerate(lst,1):
  new_lst.append(f'{y}#{x}')
```
