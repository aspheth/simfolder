```python
def divisible_by_2_and_3(lst):
  sum = 0
  for x in lst:
    if x % 2 == 0 and x % 3 == 0:
      sum += x
  return sum
```

```python
#Python ДЗ 5: Наименьший делитель
```

```python
def smallestDivisor(n):
  delit = 2
  del_lst = []
  while delit < n:
    if n % delit ==0:
      del_lst.append(delit)
    delit+=1
  if del_lst:
    return min(del_lst)
  else:
    return n
```

```python
smallestDivisor(7)
```

```python
#Python ДЗ 5: Проверка на пустые строки
```

```python
def adder(lst):
  res = []
  if lst == []:
    return 'Список пуст'
  else:
    for x in lst:
      if x == '':
        return ('Есть пустая строка', res)
      else:
        res.append(x)
  return res
```

```python
adder(lst)
```

```python
def find_n(n):
  res = 0
  count = 0
  while res < n:
    count += 1
    res += count
  return count
```

```python
find_n(4)
```

```python
def geometric_sum(n):
  num = 1
  sum = 0
  while num <= n:
    sum += 1/num
    num+=1
  return sum
```

```python
def list_numbers(start,stop):
  lst = []
  if start < stop:
    for x in range(start,stop+1):
      lst.append(x)
    return lst
  else:
    for x in range(start,stop-1,-1):
      lst.append(x)
    return lst
```

```python
my_list = [[1, 2], (3, 4), 'Пять', 'Шесть', True, False]
```

```python
def flt_lst(my_list):
    res = []
    for x in my_list:
        if isinstance(x,(list,tuple)):
            res.extend(x)
        else:
            res.append(x)   
    return res         
```

```python
flt_lst(my_list)
```

```python
#Python. ДЗ 11: Именованные и позиционные аргументы
```

```python
def args_processing(*args,**kwargs):
    print('Позиционные аргументы:')
    if args:
        for x in args:
            print(x)
    print('Именованные аргументы:')   
    if kwargs:
        for x,y in kwargs.items():
            print(f'ключ - {x}, значение - {y}')     
```

```python
args_processing(1, [1], a=[2, 3], b='s')
```

```python
args_processing([2, 3], 's')
```
