```python
data = [
    {
        'name': 'Иван',
        'income': 5000,
        'expenses': 3000,
        'date': '2023-04-01'
    },
    {
        'name': 'Мария',
        'income': 7000,
        'expenses': 6000,
        'date': '2023-04-01'
    }
]

period = '2023-04'
```

```python
filtered = list(filter(lambda x: '2023-04' in x['date'],data))
```

```python
list(map(lambda x: x['income'] - x['expenses']  ,filtered))
```

```python
max(map(lambda x: x['income'] - x['expenses']  ,list(filter(lambda x: '2023-04' in x['date'],data))))
```

```python
def client_with_largest_gap(data,period):
  return max(map(lambda x: x['income'] - x['expenses']  ,list(filter(lambda x: period in x['date'],data))))
```

```python
client_with_largest_gap(data,period)
```

```python
result = max(filter(lambda x: '2023-04' in x['date'], data),
             key=lambda x: x['income'] - x['expenses'])
```
