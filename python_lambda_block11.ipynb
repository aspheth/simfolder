```python
data = [
    {
        'name': 'Иван',
        'income': 5000,
        'expenses': 3000,
        'date': '2023-04-01'
    },
    {
        'name': 'Мария',
        'income': 7000,
        'expenses': 6000,
        'date': '2023-04-01'
    }
]

period = '2023-04'
```

```python
filtered = list(filter(lambda x: '2023-04' in x['date'],data))
```

```python
list(map(lambda x: x['income'] - x['expenses']  ,filtered))
```

```python
max(map(lambda x: x['income'] - x['expenses']  ,list(filter(lambda x: '2023-04' in x['date'],data))))
```

```python
def client_with_largest_gap(data,period):
  return max(map(lambda x: x['income'] - x['expenses']  ,list(filter(lambda x: period in x['date'],data))))
```

```python
client_with_largest_gap(data,period)
```

```python
result = max(filter(lambda x: '2023-04' in x['date'], data),
             key=lambda x: x['income'] - x['expenses'])
```

```python
#Python ДЗ 7: Общие продажи товара за несколько лет
```

```python
table = [('Календарь', '2020-01-01', 100.0),
         ('Календарь', '2021-03-15', 150.0),
         ('Часы', '2020-05-12', 80.0),
         ('Календарь', '2022-02-28', 200.0),
         ('Часы', '2021-11-19', 120.0),
         ('Блокнот', '2022-04-30', 50.0)]
```

```python
res = list(filter(lambda x: x[0] == 'Календарь' and x[1][:4] in ('2020', '2021', '2022') ,table ))
```

```python
sum(x[2] for x in res)
```
