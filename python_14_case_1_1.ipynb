```python
# РЕШЕНИЕ 1
def count_success_and_failure(file_path):
    with open('auto_purchase.log', 'r') as f:
        lines = list(f.readlines())
        err_list = []
        upgr_list = []
        res = 0
        for i, line in enumerate(lines):
            if 'Обновляем подписку' in line and i + 1 < len(lines) and 'ERROR' in lines[i + 1]:
                err_list.append((line, lines[i + 1]))
                res += 1
            else:
                upgr_list.append(line)

    return (len(list(filter(lambda x: 'payment_method_id = None\n' not in x and 'Обновляем подписку пользователю' in x, upgr_list))), res)

print(count_success_and_failure('auto_purchase.log'))
```
