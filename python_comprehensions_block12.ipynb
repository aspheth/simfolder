```python
lst = list(range(1, 21))
```

```python
lst
```

```python
max_arg = []
for x in lst:
  res_list = []
  for i in range(1,10):
    if x % i == 0:
      res_list.append(i)
  max_arg.append(max(res_list))

```

```python
max_arg
```

```python
[i for x in lst for i in range(1,10) if x % i == 0]
```

```python
res = [max(i for i in range(1,10) if x % i == 0) for x in lst]
```

```python
res
```
