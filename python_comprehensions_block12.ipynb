```python
max_arg = []
for x in lst:
  res_list = []
  for i in range(1,10):
    if x % i == 0:
      res_list.append(i)
  max_arg.append(max(res_list))

```

```python
def encryption(string):
    lst = list(string)

    for i in range(len(lst)):

        if ord(lst[i]) in range(97, 121):
            lst[i] = chr(ord(lst[i]) + 1)

    return ''.join(lst)
```
